```python
columns_range = data.loc[:, 'Natural Gas':'Waste Heat']

renewable_columns = ['Dammed Hydro', 'River', 'Wind', 'Solar', 'Geothermal', 'Biomass']
non_renewable_columns = [col for col in columns_range.columns if col not in renewable_columns]

renewable_total = data[renewable_columns].sum().sum()
non_renewable_total = data[non_renewable_columns].sum().sum()

total_mwh = data['Total (MWh)'].sum()

# YÃ¼zdeleri hesaplamak iÃ§in toplamÄ± bulalÄ±m
total_generation = renewable_total + non_renewable_total + total_mwh

# YÃ¼zdelik deÄŸerleri hesaplayalÄ±m
percentages = [
    (renewable_total / total_generation) * 100,
    (non_renewable_total / total_generation) * 100
]

labels = ['Renewable', 'Non-Renewable']

plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=labels, autopct='%1.1f%%', colors=['yellow', '#41B3A2'], startangle=140)
plt.title('Distribution of Energy Generation by Source')
plt.show()
```

```python
data['Date_Time'] = pd.to_datetime(data['Date_Time'], format='%d.%m.%Y %H:%M')

data['Date'] = data['Date_Time'].dt.date
data['Hour'] = data['Date_Time'].dt.hour

data_filtered = data[data['Date'] == pd.to_datetime('01.01.2020', format='%d.%m.%Y').date()]

plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x=data['Hour'], y='Solar', color='red')
plt.title('Solar Energy Distribution on 01.01.2020')
plt.xlabel('Hour')
plt.ylabel('Solar Energy (kWh or MWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data.columns = data.columns.str.strip()

data['Year'] = data['Date_Time'].dt.year
data['Month'] = data['Date_Time'].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

data['Season'] = data['Month'].apply(get_season)

data_2022 = data[data['Year'] == 2022]

seasonal_averages = data_2022.groupby(['Season'])[['Solar', 'River', 'Wind']].mean().reset_index()

filtered_data = seasonal_averages[seasonal_averages['Season'].isin(['Spring', 'Summer', 'Autumn', 'Winter'])]

melted_data = filtered_data.melt(id_vars='Season', var_name='Type', value_name='Average Value')

sns.catplot(
    data=melted_data,
    x='Type',
    y='Average Value',
    hue='Season',  
    kind='bar',
    height=6,
    aspect=1.5
)

plt.title('Comparison of Average Solar, River, and Wind Values in Summer and Winter 2022')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# GÃ¼nlÃ¼k tÃ¼ketime gÃ¶re yeniden Ã¶rnekleyin
data_consumption = pd.DataFrame(data['Consumption (MWh)'].resample('D').sum())

# Veriyi normalize etmek iÃ§in StandardScaler kullanacaÄŸÄ±z
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_consumption.values.reshape(-1, 1))

# Gecikmeli Ã¶zellikler oluÅŸturma fonksiyonu
def create_features(data, lag=1):
    X, y = [], []
    for i in range(lag, len(data)):
        X.append(data[i-lag:i])
        y.append(data[i])
    return np.array(X), np.array(y)

# Ã–zellikleri oluÅŸtur ve boyutlarÄ±nÄ± kontrol et
lag = 7  # Son 7 gÃ¼nÃ¼ kullanarak tahmin yapacaÄŸÄ±z
X, y = create_features(scaled_data, lag)
X = X.reshape(X.shape[0], -1)

# Veriyi eÄŸitim ve test setlerine ayÄ±r
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# KNN modelini tanÄ±mla
knn_model = KNeighborsRegressor(n_neighbors=5)

# Modeli eÄŸit
knn_model.fit(X_train, y_train)

# Test seti Ã¼zerinde tahmin yap
y_pred = knn_model.predict(X_test)

# Performans metriklerini hesapla
final_mse = mean_squared_error(y_test, y_pred)
final_rmse = np.sqrt(final_mse)
r_squared = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {final_mse}")
print(f"Root Mean Squared Error (RMSE): {final_rmse}")
print(f"R-Squared (R-Kare): {r_squared}")

# Tahmin edilen deÄŸerleri gerÃ§ek ve test setiyle birlikte grafikte gÃ¶sterelim
plt.figure(figsize=(12, 6))

dates = data_consumption.index[-len(y_test):]
plt.plot(dates, scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual Values')
plt.plot(dates, scaler.inverse_transform(y_pred.reshape(-1, 1)), label='Predicted Values', linestyle='--')

# Gelecek tahminler iÃ§in tarih aralÄ±ÄŸÄ± ekleyelim
future_preds = []
current_input = X_test[-1].reshape(1, -1)

for _ in range(30):  # 30 gÃ¼nlÃ¼k tahmin
    future_pred = knn_model.predict(current_input)
    future_preds.append(future_pred[0])
    current_input = np.roll(current_input, -1)
    current_input[0, -1] = future_pred.item()

future_dates = pd.date_range(dates[-1] + pd.Timedelta(days=1), periods=30, freq='D')
plt.plot(future_dates, scaler.inverse_transform(np.array(future_preds).reshape(-1, 1)), label='Future Predictions', linestyle='-.', color='green')

plt.title('KNN Predictions with Future Forecast')
plt.xlabel('Date')
plt.ylabel('Consumption (MWh)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.scatter(scaler.inverse_transform(y_test.reshape(-1, 1)), scaler.inverse_transform(y_pred.reshape(-1, 1)), alpha=0.5, color='blue')
plt.plot([min(scaler.inverse_transform(y_test.reshape(-1, 1))), max(scaler.inverse_transform(y_test.reshape(-1, 1)))], 
         [min(scaler.inverse_transform(y_test.reshape(-1, 1))), max(scaler.inverse_transform(y_test.reshape(-1, 1)))], 
         color='red', linestyle='--', label='Perfect Prediction Line')

plt.title('KNN Model: Predicted vs Actual Values')
plt.xlabel('Actual Consumption (MWh)')
plt.ylabel('Predicted Consumption (MWh)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

# Veriyi yÃ¼kle ve filtrele
data_consumption = pd.DataFrame(data['Consumption (MWh)'].resample('D').sum())

# Veriyi normalize etmek iÃ§in StandardScaler kullanacaÄŸÄ±z
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_consumption.values.reshape(-1, 1))

# Gecikmeli Ã¶zellikler oluÅŸturma fonksiyonu
def create_features(data, lag=1):
    X, y = [], []
    for i in range(lag, len(data)):
        X.append(data[i-lag:i])
        y.append(data[i])
    return np.array(X), np.array(y)

# Ã–zellikleri oluÅŸtur ve boyutlarÄ±nÄ± kontrol et
lag = 7  # Son 7 gÃ¼nÃ¼ kullanarak tahmin yapacaÄŸÄ±z
X, y = create_features(scaled_data, lag)
X = X.reshape(X.shape[0], -1)

# Polinomial Ã¶zellikler ekleyerek model performansÄ±nÄ± artÄ±rmak
poly = PolynomialFeatures(degree=2)  # Polinomial derecesini artÄ±rarak daha karmaÅŸÄ±k bir model oluÅŸturabiliriz
X_poly = poly.fit_transform(X)

# Veriyi eÄŸitim ve test setlerine ayÄ±r
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, shuffle=False)

# Ridge modelini oluÅŸtur ve GridSearchCV ile en iyi parametreleri bul
param_grid = {'alpha': np.logspace(-4, 4, 20)}
ridge_model = GridSearchCV(Ridge(), param_grid, cv=5, scoring='neg_mean_squared_error')
ridge_model.fit(X_train, y_train)

# En iyi Ridge modelini seÃ§ ve test seti Ã¼zerinde tahmin yap
best_ridge = ridge_model.best_estimator_
y_pred = best_ridge.predict(X_test)

# Performans metriklerini hesaplayalÄ±m
final_mse = mean_squared_error(y_test, y_pred)
final_rmse = np.sqrt(final_mse)
r_squared = r2_score(y_test, y_pred)

print(f"En iyi alpha deÄŸeri: {ridge_model.best_params_['alpha']}")
print(f"Mean Squared Error (MSE): {final_mse}")
print(f"Root Mean Squared Error (RMSE): {final_rmse}")
print(f"R-Squared (R-Kare): {r_squared}")

# Tahmin edilen deÄŸerleri gerÃ§ek ve test setiyle birlikte grafikte gÃ¶sterelim
plt.figure(figsize=(12, 6))

dates = data_consumption.index[-len(y_test):]
plt.plot(dates, scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual Values')
plt.plot(dates, scaler.inverse_transform(y_pred.reshape(-1, 1)), label='Predicted Values', linestyle='--')

# Gelecek tahminler iÃ§in tarih aralÄ±ÄŸÄ± ekleyelim
future_preds = []
current_input = X_test[-1].reshape(1, -1)

for _ in range(30):  # 30 gÃ¼nlÃ¼k tahmin
    future_pred = best_ridge.predict(current_input)
    future_preds.append(future_pred[0])
    current_input = np.roll(current_input, -1)
    current_input[0, -1] = future_pred.item()

# Gelecek tahminleri Ã§iz
future_dates = pd.date_range(dates[-1] + pd.Timedelta(days=1), periods=30, freq='D')
plt.plot(future_dates, scaler.inverse_transform(np.array(future_preds).reshape(-1, 1)), label='Future Predictions', linestyle='-.', color='green')

plt.title(f'Ridge Predictions with Future Forecast (Best Alpha={ridge_model.best_params_["alpha"]})')
plt.xlabel('Date')
plt.ylabel('Consumption (MWh)')
plt.legend()
plt.show()

# Scatter plot oluÅŸturma
plt.figure(figsize=(12, 6))

# GerÃ§ek deÄŸerler ve tahmin edilen deÄŸerleri scatter plot Ã¼zerinde gÃ¶sterelim
plt.scatter(scaler.inverse_transform(y_test.reshape(-1, 1)), scaler.inverse_transform(y_pred.reshape(-1, 1)), color='blue', alpha=0.6)
plt.plot([min(scaler.inverse_transform(y_test.reshape(-1, 1))), max(scaler.inverse_transform(y_test.reshape(-1, 1)))],
         [min(scaler.inverse_transform(y_test.reshape(-1, 1))), max(scaler.inverse_transform(y_test.reshape(-1, 1)))],
         color='red', linestyle='--', label='Perfect Prediction Line')

plt.title('Ridge Model: Predicted vs Actual Values')
plt.xlabel('Actual Consumption (MWh)')
plt.ylabel('Predicted Consumption (MWh)')
plt.legend()
plt.grid(True)
plt.show()
```
